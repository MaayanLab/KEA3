```python
#Create an empty list 'l' representing the links
links = list()

for index, rowData in HPRD.iterrows():
    pair = {"source": rowData['kinase_organism'], 
            "target":rowData['substrate']}
    links.append(pair)

links
```

```python
kinases = list(HPRD.kinase_organism.unique())
kinases
```

```python
targets = list(HPRD.substrate.unique())
targets
```

```python
tar = dict((key, "target") for key in targets)
kin = dict((key, "kinase") for key in kinases)
```

```python
tar
```

```python
kin
```

```python
#create an empty list 'nodes' representing the nodes
nodes = list()

for key, value in kin.items():
    pair = {'id':key, 'group':value}
    nodes.append(pair)

for key, value in tar.items():
    pair = {'id':key, 'group':value}
    nodes.append(pair)
    
nodes
```

```python
combined = OrderedDict({"nodes": nodes, "links": links})
```
